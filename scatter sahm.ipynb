## The amount of profit and loss of Google and Microsoft shares

```python
import yfinance as yf
```

```python
import pandas as pd
```

```python
df=yf.download('GOOG MSFT',period='50d')
```

```python
df_gm=df['Adj Close']
```

```python
df_gm.pct_change().plot(figsize=(16,6) , kind="bar",grid=True)
```

```python
#for google pct_chance positive
list1=['2023-08-31','2023-09-07','2023-09-18','2023-10-17',]
for i in list1:
    print(df_gm.loc[i][0])
```

```python
#for microsoft pct_chance negative
list1=['2023-08-31','2023-09-07','2023-09-18','2023-10-17']
for i in list1:
    print(df_gm.loc[i][1])
```

```python
#for microsoft pct_chance positive
list1=['2023-09-01','2023-09-05','2023-09-29','2023-10-05','2023-10-19','2023-10-25']
for i in list1:
    print(df_gm.loc[i][0])
```

```python
#for google pct_chance negative
list1=['2023-09-01','2023-09-05','2023-09-29','2023-10-05','2023-10-19','2023-10-25']
for i in list1:
    print(df_gm.loc[i][1])
```

### Relationships between two stocks

```python
df_gm.pct_change().plot(x="MSFT",y="GOOG",figsize=(16,6) , kind="scatter",grid=True,colorbar=True)
```

 Result:There is a perfect positive correlation between the two stocks

### Relations between the two shares of Iran Khodro and Saipa

```python
import tsemodule5 as ts5
```

```python
sipa=ts5.stock('خساپا',150,standard=True)['Close']
```

```python
irkh=ts5.stock('خودرو',150,standard=True)['Close']
```

```python
sipa=sipa.loc['2023-03-20':'2023-07-01']
irkh=irkh.loc['2023-03-20':'2023-07-01']
```

```python
dfp=pd.DataFrame({'Khaspa':sipa,'Khodro':irkh})
```

```python
dfp
```

```python
sipa.pct_change().plot(figsize=(16,6) ,grid=True)
```

```python
irkh.pct_change().plot(figsize=(16,6) ,grid=True)
```

```python
dfp.plot(x='Khodro',y='Khaspa',figsize=(16,6),title='scatter plot from two stock',kind='scatter',colorbar=True,grid=True)
```

```python
dfp.plot(x='Khodro', y='Khaspa', figsize=(16, 6), kind='scatter', c=dfp['Khodro'], colorbar=True, grid=True)
```

Result:

About 60 percent of the data are between 0.2 and 0.8, or rather, 20 to 80 percent of the data and there is a lot of dispersion in this plot.

There is a weak positive correlation between the two stocks and there is a gap between the data indicating an unexpected gap in the data.
